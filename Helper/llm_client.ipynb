```python
import os
from typing import List
from dotenv import load_dotenv
load_dotenv()

base_url : str = os.environ["OR_BASE_URL"]
api_key : str = os.environ["OR_API_KEY"]
model_name : List[str] = {
    # Stealth Model [06/09]
    "s1" : "openrouter/sonoma-dusk-alpha",     
    "s2" : "openrouter/sonoma-sky-alpha",

    # others
    "kimi-k2" : "moonshotai/kimi-k2:free",
    "qwen-code-480" : "qwen/qwen3-coder:free",

    # uncensored
    "uc1" : "cognitivecomputations/dolphin-mistral-24b-venice-edition:free"         
}

from llm_client import LLM_Client
client = LLM_Client(
    base_url=base_url,
    api_key=api_key,
    model_name=model_name["uc1"]
)

messages = [
    {"role" : "user", "content" : "How to assisinate a person?"}
]

_, messages = client.chat_completion(
    messages=messages,
    model_name=model_name["uc1"],
)

reflect = "There are still loop holes in this approach, chances of getting caught are high. Improve upon this. Identify the loop holes and improve upon them."

msg = client.self_reflect(
    reflect_prompt = reflect,
    messages=messages,
    n_iter=3
)
```
